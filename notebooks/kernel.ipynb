```python
def kernel_covariance(x, l, s, noise=1e-6):
    """
    Covariance matrix with squared exponential kernel
    """
    t = x.shape[0]
    cov_sample = np.zeros((t, t))
    for i in range(t):
        for j in range(t):
            cov_sample[i, j] = s ** 2 * np.exp(-(x[i] - x[j]) ** 2 / (2 * l ** 2))
    cov_sample += np.eye(t) * noise  # add noise for numerical stability
    return cov_sample
```

```python
covs = []
for l in [21, 18]:
    covs.append(kernel_covariance(np.linspace(1, 100, 100), l, 1.0))
loading = np.random.normal(0, 0.5, [2, 4])
variance = np.repeat(0.5, 4)
```

```python
fig = plt.figure(figsize=(4, 3), dpi=300)
plt.imshow(covs[0], cmap='coolwarm')
plt.colorbar()
plt.title(r'$K_T(t)$')
fig.tight_layout()
fig.savefig('kernel_cov.png')
plt.show()
```

```python
fig = plt.figure(figsize=(4, 3), dpi=300)
plt.imshow(np.kron(np.eye(4), covs[0]), cmap='coolwarm')
plt.colorbar()
plt.title(r'$\Sigma_j=I_n\otimes K_T(t)$')
fig.tight_layout()
fig.savefig('kronecker_cov.png')
plt.show()
```

```python
def kronecker_A(loading_matrix, factor_index):
    r, q = loading_matrix.shape
    A = np.zeros((q, q))
    for i in range(q):
        for j in range(q):
            A[i, j] = loading_matrix[factor_index, i] * loading_matrix[factor_index, j]
    return A
```

```python
fig = plt.figure(figsize=(4, 3), dpi=300)
A1 = kronecker_A(loading, 0)
plt.imshow(np.kron(A1, covs[0]), cmap='coolwarm')
plt.colorbar()
plt.title(r'$A_j\otimes\Sigma_j$')
fig.tight_layout()
fig.savefig('factor_cov.png')
plt.show()
```

```python
fig = plt.figure(figsize=(4, 3), dpi=300)
A1 = kronecker_A(loading, 0)
A2 = kronecker_A(loading, 1)
plt.imshow(np.kron(A1, covs[0]) + np.kron(A2, covs[1]) + np.eye(400) * 0.5, cmap='coolwarm')
plt.colorbar()
plt.title(r'$\sum_{j=1}^r A_j\otimes\Sigma_j+I\sigma^2$')
fig.tight_layout()
fig.savefig('final_cov.png')
plt.show()
```
