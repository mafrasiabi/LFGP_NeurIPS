```python
from tqdm import tqdm as tqdm
```

```python
cov_series = np.zeros((259, 40, 10, 10))  
log_series = np.zeros((259, 40, 55))  
for i in tqdm(range(259)):
    for t in range(40):
        window = np.transpose(cups_data[i, :, (t * 5):(t * 5 + 30)])
        cov = np.cov(window, rowvar=False)
        cov_series[i, t, :, :] = cov
        utv = logm(cov)[np.triu_indices(10)]
        log_series[i, t, :] = utv
```

```python
def create_symmetric_matrix(upper, n):
    mat = np.zeros((n, n))
    k = 0
    for i in range(n):
        mat[i, i:] = upper[k:(k + n - i)]
        k += n - i
    for i in range(1, n):
        for j in range(i):
            mat[i, j] = mat[j, i]
    return mat
```

```python
cor_hat = np.zeros((10, 10, 10))
for i in range(10):
    cor_hat[i, :, :] = expm(create_symmetric_matrix(hat[i * 4, :] + np.mean(all_vector, axis=0)[j], 10))
```

```python
cor_true = np.zeros((10, 10, 10))
for i in range(10):
    cor_true[i, :, :] = expm(create_symmetric_matrix(true[i * 4, :] + np.mean(all_vector, axis=0)[j], 10))
```

```python
fig = plt.figure(figsize=(20, 8), dpi=200)
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(cor_hat[i, :, :])
    plt.xticks([])
    plt.yticks([])
    plt.colorbar()
plt.show()
```
