```python
import numpy as np
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['font.family'] = 'serif'
```

```python
gp = GaussianProcessRegressor(kernel=RBF(length_scale=10))
```

```python
x = np.linspace(1, 100, 100)
curves = gp.sample_y(x.reshape((100, 1)), 4)
F = curves
```

```python
print(F.shape)
```

```python
loading = np.random.normal(0, 2, [4, 10])
```

```python
loading[2:4, :] = 0
```

```python
fig = plt.figure(figsize=(5, 2), dpi=150)
plt.imshow(np.absolute(loading), cmap='hot')
plt.colorbar()
plt.show()
```

```python
Y_true = np.matmul(F, loading)
Y = Y_true + np.random.normal(0, 0.5, [100, 10])
```

```python
from blr import sample_regression_posterior
```

```python
loading_sample = np.zeros((500, 4, 10))
for i in range(500):
    loading_draw, _ = sample_regression_posterior(Y, F, loading_prior_params=[0, 0.1], variance_prior_params=[1, 1])
    loading_sample[i, :, :] = loading_draw
```

```python
loading_hat = np.median(loading_sample, axis=0)
```

```python
fig = plt.figure(figsize=(5, 2), dpi=150)
plt.imshow(np.absolute(loading_hat), cmap='hot')
plt.colorbar()
plt.show()
```

```python
from factorgp import FactorGP
from inference import run_gibbs
```

```python
dims = [1, 100, 10, 4]  # n, t, q, r
model = FactorGP(dims, mu_var=[0, 0.01])
model.F = F.copy()
```

```python
results = run_gibbs(Y, model, 100, 10, 0.5, verbose=True)
```

```python
F_sample, loading_sample, variance_sample, theta_sample, traces_hist, mse_history = results
```

```python
sparse_loading = model.loading.copy()
sparse_loading[1:3, :] = 0
sparse_hat = np.matmul(model.F, sparse_loading)
print(sparse_hat.shape)
```

```python
print(np.mean((model.predict() - Y) ** 2))
print(np.mean((sparse_hat - Y) ** 2))
```

```python
fig = plt.figure(figsize=(4, 3), dpi=150)
plt.plot(mse_history)
plt.title('4 Factor Model MSE')
plt.xlabel('Iteration')
plt.show()
```
