```python
import numpy as np
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['font.family'] = 'serif'
```

```python
sim_data = np.load('sim_data_new.npy')
```

```python
def prepare_data(sim_data, scenario, data_idx):
    length = 500
    start = data_idx * length * 4 + scenario * length
    end = start + length
    X = np.asarray(sim_data[start:end, :10])
    Y = np.asarray(sim_data[start:end, 10:65])
    F = np.asarray(sim_data[start:end, 65:69])
    return X, Y, F
```

```python
X, Y, F = prepare_data(sim_data, 0, 0)
plt.plot(F[:, 0])
plt.plot(F[:, 1])
plt.plot(F[:, 2])
plt.plot(F[:, 3])
plt.show()
```

```python
from hmmlearn import hmm
```

```python
model = hmm.GaussianHMM(n_components=5, covariance_type='full', n_iter=50, algorithm='map').fit(X)
```

```python
Z = model.predict(X)
```

```python
plt.plot(Z)
plt.show()
```

```python
from scipy.linalg import logm
```

```python
def predict_cov(Z, covs):
    """
    Predict covariance covariance Log-Euclidean vector time series.
    
    Args
        Z: (numpy array) predicted discrete latent states of length t
        covs: (numpy array) [k, n, n] covariance matrices for the k latent states
    """
    t = Z.shape[0]
    n = covs[0].shape[0]
    #cov_hat = np.zeros((t, n, n))
    log_hat = np.zeros((t, int(0.5 * n * (n + 1))))
    for i in range(t):
        latent = Z[i]
        #cov_hat[i, :, :] = covs[latent, :, :]
        log_hat[i, :] = logm(covs[latent, :, :])[np.triu_indices(n)]
    return log_hat
```

```python
Y_hat = predict_cov(Z, model.covars_)
```

```python
print(np.mean((Y_hat - Y) ** 2))
```

```python
def HMM(X, Y, k=4):
    model = hmm.GaussianHMM(n_components=k, covariance_type='full', n_iter=50, algorithm='map').fit(X)
    Z = model.predict(X)
    Y_hat = predict_cov(Z, model.covars_)
    return Z, Y_hat
```

```python
square_mse = np.zeros(50)
for i in range(50):
    X, Y, F = prepare_data(sim_data, 0, i)
    try:
        Z, Y_hat = HMM(X, Y, k=8)
        plt.plot(Z)
        plt.show()
        mse = np.mean((Y_hat - Y) ** 2)
        square_mse[i] = mse
        print(mse)
    except:
        square_mse[i] = -1
        print('Chicken warning')
        print(i)
        continue
```

```python
print(np.median(square_mse[square_mse > 0]))
print(np.mean(square_mse[square_mse > 0]))
print(np.std(square_mse[square_mse > 0]))
```

```python
print(np.median(piecewise_mse[piecewise_mse > 0]))
print(np.mean(piecewise_mse[piecewise_mse > 0]))
print(np.std(piecewise_mse[piecewise_mse > 0]))
```
