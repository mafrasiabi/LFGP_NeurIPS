### Hidden markov model

```python
from hmmlearn import hmm
```

```python
model = hmm.GaussianHMM(n_components=5, covariance_type='full', n_iter=100, algorithm='map').fit(X)
```

```python
Z = model.predict(X)
```

```python
plt.plot(Z)
plt.show()
```

```python
def predict_cov(Z, covs):
    """
    Predict covariance covariance Log-Euclidean vector time series.
    
    Args
        Z: (numpy array) predicted discrete latent states of length t
        covs: (numpy array) [k, n, n] covariance matrices for the k latent states
    """
    t = Z.shape[0]
    n = covs[0].shape[0]
    #cov_hat = np.zeros((t, n, n))
    log_hat = np.zeros((t, int(0.5 * n * (n + 1))))
    for i in range(t):
        latent = Z[i]
        #cov_hat[i, :, :] = covs[latent, :, :]
        log_hat[i, :] = logm(covs[latent, :, :])[np.triu_indices(n)]
    return log_hat
```

```python
Y_hat = predict_cov(Z, model.covars_)
```

```python
for j in range(15):
    plt.plot(Y_hat[:, j])
plt.show()
```

```python
print(np.mean((Y_hat - Y) ** 2))
```

```python
def prepare_data(sim_data, scenario, data_idx):
    length = 1000
    sample_size = 10
    start = data_idx * length * 4 + scenario * length
    end = start + length
    X = np.asarray(sim_data[start:end, :5])
    Y = np.asarray(sim_data[start:end, 5:20])
    F = np.asarray(sim_data[start:end, 20:22])
    return X, Y, F
```

```python
def HMM(X, Y, k=4):
    model = hmm.GaussianHMM(n_components=k, covariance_type='full', n_iter=50, algorithm='map').fit(X)
    Z = model.predict(X)
    Y_hat = predict_cov(Z, model.covars_)
    return Z, Y_hat
```

```python
all_mse = np.zeros(50)
for i in range(50):
    X, Y, F = prepare_data(sim_data, 3, i)
    plt.plot(F[:, 0])
    plt.plot(F[:, 1])
    plt.show()
    try:
        Z, Y_hat = HMM(X, Y, k=4)
        plt.plot(Z)
        plt.show()
        mse = np.mean((Y_hat - Y) ** 2)
        all_mse[i] = mse
        print(mse)
    except:
        continue
```

```python
plt.hist(all_mse)
plt.show()
```

```python
print(np.mean(all_mse))
```
