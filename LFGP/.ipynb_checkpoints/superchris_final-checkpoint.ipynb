```python
import numpy as np
```

```python
from scipy.io import loadmat
from scipy.linalg import expm, logm
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_path = '/Users/linggeli/neuroscience/data/SuperChris/super_chris_extraction_odor2s.mat'
#data_path = '/Users/linggeli/neuroscience/data/Stella/stella_extraction_odor2s.mat'
data_odor = loadmat(data_path)
```

```python
trial_info = data_odor['trialInfo']
```

```python
select_odor_B = (trial_info[:, 0] == 1) & (trial_info[:, 1] == 1) & (trial_info[:, 3] == 2)
select_odor_C = (trial_info[:, 0] == 1) & (trial_info[:, 1] == 1) & (trial_info[:, 3] == 3)
```

```python
print(data_odor['lfpEpoch'].shape)
```

```python
lfp_odor_B = data_odor['lfpEpoch'][select_odor_B, 2000:4000, 5:10]
lfp_odor_C = data_odor['lfpEpoch'][select_odor_C, 2000:4000, 5:10]
```

```python
print(lfp_odor_B.shape)
print(lfp_odor_C.shape)
```

```python
cov_series_odor_B = np.zeros((40, 50, 5, 5))  
utv_series_odor_B = np.zeros((40, 50, 15))  
for i in range(40):
    for t in range(50):
        lfp_window = lfp_odor_B[i, (t * 20):(t * 20 + 100), :]
        cov = np.cov(lfp_window, rowvar=False)
        cov_series_odor_B[i, t, :, :] = cov
        utv = logm(cov)[np.triu_indices(5)]
        utv_series_odor_B[i, t, :] = utv
```

```python
cov_series_odor_C = np.zeros((30, 50, 5, 5))  
utv_series_odor_C = np.zeros((30, 50, 15))  
for i in range(30):
    for t in range(50):
        lfp_window = lfp_odor_C[i, (t * 20):(t * 20 + 100), :]
        cov = np.cov(lfp_window, rowvar=False)
        cov_series_odor_C[i, t, :, :] = cov
        utv = logm(cov)[np.triu_indices(5)]
        utv_series_odor_C[i, t, :] = utv
```

```python
data_list = []
for i in range(20):
    data_list.append(utv_series_odor_B[i, :, :])
for i in range(20):
    data_list.append(utv_series_odor_C[i, :, :])
```

```python
all_data = np.concatenate(data_list)
Y = all_data - np.mean(all_data, axis=0)
#Y = Y / np.std(all_data, axis=0)
print(Y.shape)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=4)
components = pca.fit_transform(Y)
```

```python
Y_hat = pca.inverse_transform(components)
mse = np.mean((Y - Y_hat) ** 2)
print(mse)
```

```python
print(pca.explained_variance_ratio_)
```

```python
from factorgp import FactorGP
from inference import run_gibbs
```

```python
n = 40
t = 50
q = 15
r = 2
```

```python
dims = [n, t, q, r]
model = FactorGP(dims, mu_var=[0, 0.1], gamma=[5, 1])
```

```python
results = run_gibbs(Y, model, 500, 10, 0.1, verbose=True)
```
